## Replacing nulls in meaneduc,SQBmeaned by average in their columns


```python
train_df['meaneduc'].fillna((train_df['meaneduc'].mean()), inplace=True)
train_df['SQBmeaned'].fillna((train_df['SQBmeaned'].mean()), inplace=True)
```

## Dealing with columns dependency,edjefe,edjefa


```python
train_df['edjefa'][train_df['edjefa']=='yes']=train_df['edjefa'][~train_df['edjefa'].isin(['yes','no'])].astype(float).mean()
```

```python
train_df['edjefa'][train_df['edjefa']=='no']=0
```

```python
train_df['edjefe'][train_df['edjefe']=='yes']=train_df['edjefe'][~train_df['edjefe'].isin(['yes','no'])].astype(float).mean()
```

```python
#train_df['edjefa'][~train_df['edjefa'].isin(['yes','no'])].astype(float).mean()
train_df['edjefa'].unique()
```

```python
train_df['edjefe'][train_df['edjefe']=='no']=0
```

```python
train_df['dependency'][~train_df['dependency'].isin(['yes','no'])].astype(float).mean()
```

```python
train_df['dependency'][train_df['dependency']=='yes']=train_df['dependency'][~train_df['dependency'].isin(['yes','no'])].astype(float).mean()
```

```python
train_df['dependency'][train_df['dependency']=='no']=0
```

```python
import plotly.subplots as tools
#Tablet
tr1=train_df.groupby(['Target'])["v18q"].value_counts().to_frame()
tr1.columns=['count']
tr1['Target'] = tr1.index.get_level_values('Target')
tr1['v18q'] = tr1.index.get_level_values('v18q')

#Refrig
tr2=train_df.groupby(['Target'])["refrig"].value_counts().to_frame()
tr2.columns=['count']
tr2['Target'] = tr2.index.get_level_values('Target')
tr2['refrig'] = tr2.index.get_level_values('refrig')
#levels = ["NonVulnerable", "Moderate Poverty", "Vulnerable", "Extereme Poverty"]

#Computer 
tr3=train_df.groupby(['Target'])["computer"].value_counts().to_frame()
tr3.columns=['count']
tr3['Target'] = tr3.index.get_level_values('Target')
tr3['computer'] = tr3.index.get_level_values('computer')

#television
tr4=train_df.groupby(['Target'])["television"].value_counts().to_frame()
tr4.columns=['count']
tr4['Target'] = tr4.index.get_level_values('Target')
tr4['television'] = tr4.index.get_level_values('television')

#mobilephone
tr5=train_df.groupby(['Target'])["mobilephone"].value_counts().to_frame()
tr5.columns=['count']
tr5['Target'] = tr5.index.get_level_values('Target')
tr5['mobilephone'] = tr5.index.get_level_values('mobilephone')


plot1=go.Bar(y=tr1[tr1['v18q']==1]['count'], x=tr1[tr1['v18q']==1]['Target'],marker=dict(color="green", opacity=0.6))
plot2=go.Bar(y=tr1[tr1['v18q']==0]['count'], x=tr1[tr1['v18q']==0]['Target'],marker=dict(color="red", opacity=0.6))
plot3=go.Bar(y=tr2[tr2['refrig']==1]['count'], x=tr2[tr2['refrig']==1]['Target'],marker=dict(color="green", opacity=0.6))
plot4=go.Bar(y=tr2[tr2['refrig']==0]['count'], x=tr2[tr2['refrig']==0]['Target'],marker=dict(color="red", opacity=0.6))
plot5=go.Bar(y=tr3[tr3['computer']==1]['count'], x=tr3[tr3['computer']==1]['Target'],marker=dict(color="green", opacity=0.6))
plot6=go.Bar(y=tr3[tr3['computer']==0]['count'], x=tr3[tr3['computer']==0]['Target'],marker=dict(color="red", opacity=0.6))
plot7=go.Bar(y=tr4[tr4['television']==1]['count'], x=tr4[tr4['television']==1]['Target'],marker=dict(color="green", opacity=0.6))
plot8=go.Bar(y=tr4[tr4['television']==0]['count'], x=tr4[tr4['television']==0]['Target'],marker=dict(color="red", opacity=0.6))
plot9=go.Bar(y=tr5[tr5['mobilephone']==1]['count'], x=tr5[tr5['mobilephone']==1]['Target'],marker=dict(color="green", opacity=0.6))
plot10=go.Bar(y=tr5[tr5['mobilephone']==0]['count'], x=tr5[tr5['mobilephone']==0]['Target'],marker=dict(color="red", opacity=0.6))


titles = ["Tablet", "Refrigirator", "Computer", "Television", "MobilePhone"]
fig = tools.make_subplots(rows=3, cols=2, print_grid=False, subplot_titles=titles)
fig.append_trace(plot1, 1, 1)
fig.append_trace(plot2, 1, 1)
fig.append_trace(plot3, 1, 2)
fig.append_trace(plot4, 1, 2)
fig.append_trace(plot5, 2, 1)
fig.append_trace(plot6, 2, 1)
fig.append_trace(plot7, 2, 2)
fig.append_trace(plot8, 2, 2)
fig.append_trace(plot9, 3, 1)
fig.append_trace(plot10, 3, 1)


fig['layout'].update(height=1000, title="What do Households Own", barmode="stack", showlegend=False)
iplot(fig)

```
